# Logistic Population growth simulation

```python
import matplotlib.pyplot as plt

initial_population = 1
growth_rate = 0.5
hours = 24
maximum_capacity = 1000
population = [initial_population]

def grow(initial):
    return initial + (growth_rate * initial * (1 - (initial / maximum_capacity)))  # Formula for Logistic growth

# Simulate the growth over the specified number of hours
for i in range(1, hours):
    population.append(grow(population[-1]))

# Plot the graph of the population growth
plt.plot(range(1, hours + 1), population)
plt.title("Population Growth Over Time")
plt.xlabel("Time (hours)")
plt.ylabel("Population Count")
```
